```python
for c in df.columns:
    miss = df[c].isnull().sum()
    print("{} has {} missing value(s)".format(c,miss))
```

```python
num_cols = ['Views', 'Likes', 'Comments', 'Stream']
df[num_cols] = df[num_cols].fillna(0)
```

```python
bool_cols = ['Licensed', 'official_video']
df[bool_cols] = df[bool_cols].fillna(False)
df
```

```python
df['Views'] = df['Views'].fillna(0).astype(int)
df1 = df.sample(frac = 0.01)
```

```python
plt.boxplot(df1['Views'])
```

```python
df1['Views'].hist(bins = 30)
```

```python
low_IQR = df1['Views'].quantile(0.75) - df1['Views'].quantile(0.25)
low_limit = df1['Views'].quantile(0.25) - 1.5 * low_IQR

high_IQR = df1['Views'].quantile(0.75) - df1['Views'].quantile(0.25)
high_limit = df1['Views'].quantile(0.75) + 1.5 * high_IQR

df1['Views'] = np.where(df1['Views'] < low_limit, low_limit, df1['Views'])
df1['Views'] = np.where(df1['Views'] > high_limit, high_limit, df1['Views'])
```
